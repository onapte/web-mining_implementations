# Girvan-Newman algorithm

Example graph:  



## Import libraries

```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms.community.centrality import girvan_newman
```

## Create graph

```python
G = nx.Graph()
```

### - Add nodes

```python
G.add_nodes_from([1, 2, 3, 4, 5, 6, 7, 8, 9])
```

```python
nx.draw(G, with_labels=True, node_color='red', node_size=2000)
plt.margins(0.2)
plt.show()
```

### - Add edges

```python
G.add_edges_from([
    (1, 2), (1, 3), (1, 4), (2, 3), (3, 4), (4, 5), (4, 6),
    (5, 6), (5, 7), (5, 8), (6, 7), (6, 8), (7, 8), (7, 9)
    ])
```

### - Define coordinates for each node

```python
pos = {
    1: (9.25, 2.5),
    2: (8, 0.75),
    3: (8.05, 4),
    4: (6.5, 2.5),
    5: (4.25, 3.45),
    6: (4.1, 1.67),
    7: (2.4, 3.6),
    8: (2.1, 1.64),
    9: (0.65, 2.5)
}
```

## Display the graph

```python
nx.draw(G, pos=pos, with_labels=True, node_color='red', node_size=2000)
plt.margins(0.2)
plt.show()
```

## Use the Girvan-Newman algorithm

```python
communities = girvan_newman(G)
```

```python
node_groups = []
for com in next(communities):
    node_groups.append(list(com))

print(node_groups)
```

```python
color_map = []
for node in G:
    if node in node_groups[0]:
        color_map.append("blue")
    else:
        color_map.append("violet")
```

## Display the new graph partitioned by communities

```python
nx.draw(G, pos=pos, node_color=color_map, with_labels=True, node_size=2000)
plt.margins(0.2)
plt.show()
```
